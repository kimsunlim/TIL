```python
#barplot
#스타일 테마를 설정5가지(darkgrid,whitegrid,dark,white,ticks)
sns.set_style('whitegrid')

#그래프 객체 생성

fig= plt.figure(figsize=(15,5))
area01.set_title('titanic survived -sex')
area01=fig.add_subplot(1,3,1)
area02=fig.add_subplot(1,3,2)
area03=fig.add_subplot(1,3,3)


# 성별에 따른 생존률


sns.barplot(x='sex',y='survived',data=titanic, ax=area01)

# hue
sns.barplot(x='sex',y='survived',hue='class',data=titanic ,ax=area02)

#dodge
sns.barplot(x='sex',y='survived',hue='class',dodge=False ,data=titanic, ax=area03)


plt.show()
```

- countplot():카테고리 값별로 데이터 빈도를 표시하는 차트

```python
#barplot
#스타일 테마를 설정5가지(darkgrid,whitegrid,dark,white,ticks)
sns.set_style('darkgrid')

#그래프 객체 생성

fig= plt.figure(figsize=(15,5))

area01=fig.add_subplot(1,3,1)
area01.set_title('titanic class')

area02=fig.add_subplot(1,3,2)
area02.set_title('titanic class -who')

area03=fig.add_subplot(1,3,3)
area03.set_title('titanic class -who(stacked)')




# 기본

sns.countplot(x='class',data=titanic, ax=area01)


#hue
sns.countplot(x='class',hue='who',palette='Set2',data=titanic ,ax=area02)

#dodge
sns.countplot(x='class',hue='who',palette='Set2',dodge=False ,data=titanic, ax=area03)


```

```python
#tips를 이용한 countplot()
#요일별로 팁을 준 회수를 시각화 한다면?


sns.countplot(x='day',data=tips)
plt.title('요일별 전체팁')
plt.show()
```

- stripplot():스카터플롯처럼 모든데이터를 점으로 시각화
- jitter=True :설정하면 위치를 무작위로 바꿔서 겹치지 않게 해줌.

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
plt.title('요일별 전체팁')
plt.show()
```

```python
#점이 겹치면 좌우로 퍼짐
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.title('요일별 전체팁')
plt.show()
```

```python
#타이타익 데이터를 이용하여
# 서브플롯을 2개 영역으로 만들고
#class별 나이의 분포를 확인하가(stripplot(),swarmplot())



sns.set_style('whitegrid')

#그래프 객체 생성

fig= plt.figure(figsize=(15,5))


area01=fig.add_subplot(1,2,1)
area01.set_title('strip')


area02=fig.add_subplot(1,2,2)
area02.set_title('swarm')






sns.stripplot(x='class',y='age',data=titanic, palette='Set2', ax=area01,jitter=True)

# hue
sns.swarmplot(x='class',y='age',data=titanic, palette='Set2' ,ax=area02)


plt.show()
```

```python
# 만약에 tips에서
#요일별,성별 전체팁 stripplot()시가고하

sns.stripplot(x='day', y='total_bill', hue='sex', data=tips)
plt.title('요일별 전체 팁')
plt.legend(loc='best')
plt.show()
```

- boxplot() ,violinplot()


```python

sns.set_style('whitegrid')

#그래프 객체 생성

fig= plt.figure(figsize=(15,5))


ax01=fig.add_subplot(2,2,1)
ax01.set_title('box')


ax02=fig.add_subplot(2,2,2)
ax02.set_title('box')


ax03=fig.add_subplot(2,2,3)
ax03.set_title('violin')


ax04=fig.add_subplot(2,2,4)
ax04.set_title('violin')


sns.boxplot(x='alive',y='age',data=titanic,ax=ax01)
sns.boxplot(x='alive',y='age',data=titanic, hue='sex',ax=ax02)

sns.violinplot(x='alive',y='age',data=titanic,ax=ax03)
sns.violinplot((x='alive',y='age',data=titanic, hue='sex',ax=ax04)


plt.show()
```

- heatmap :하나 이상의 카테고리 값의 변화를 색깔 변화로 시각화 하는 차트

```python
#titanic
#피벗테이블로 범주형 변수를 각각 행,열로 재구분하여 정리
#인덱스 성별,열클래스

table=titanic.pivot_table(index=['sex'],columns=['class'],aggfunc='size')  #그룹화시킨것
#table.head()
sns.heatmap(table,
           annot=True,fmt='d',linewidth=1)  #annot:박스에 데이터 출력  fmt:정수값  linewidth:구분선
plt.show()
```
