## 서울지역 초등학교


```python
seoul_school_df= pd.read_excel('./data/elem_school_folium.xlsx')
seoul_school_df.head()
```

```python
#Feature Selection
# 학교명,학교급구분, 소재지도로명,위도,경도
 
feature_sel_df= seoul_school_df.filter(['학교명','학교급구분','소재지도로명주소','위도','경도'])
feature_sel_df.head()
```

```python
#소재지 도로명주소가 서울을 포함하고 학교급구분 초등학교만 추출

#sql like
#feature_sel_df['소재지도로명주소'].str.contains('서울특별시')
#조건에 부합하는 결과를 데이터 프레임으로 만들기 위해서는  query


#결측값 확인,제거

print(feature_sel_df.isnull().sum())
feature_sel_df.dropna(inplace=True)
print("*"*50)
print(feature_sel_df.isnull().sum())


#feature_sel_df[feature_sel_df['소재지도로명주소'].str.contains('서울특별시')]
```

```python
#조건에 부합하는 결과를 데이터 프레임으로 만들기 위해서 query()

#feature_sel_df.query("학교급구분 =='초등학교'and 소재지 도로명 주소.str.contrains('서울 특별시')")
#인덱스 추출


index_seoul =feature_sel_df['소재지도로명주소'].str.contains('서울특별시')

index_ele =feature_sel_df['학교급구분']=='초등학교'
print(len(index_seoul))
print(len(index_ele))


result_df =feature_sel_df.loc[index_seoul & index_ele]
result_df.head()
```

```python
#지도시각화

for name, lat, lng in zip(result_df.index , result_df.위도, result_df.경도) :
    g.CircleMarker([lat, lng], 
                   popup=name, 
                   icon=g.Icon(color='red'),
                  radius=10,
                  fill=True,
                  fill_color='coral',
                  fill_opacity=0.3).add_to(seoul_map)

seoul_map


```

```python
import pandas as pd
import numpy  as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
import json
%matplotlib inline
```

```python
pop_df = pd.read_excel('./data/경기도인구데이터.xlsx',index_col='구분')
pop_df.head()
```

```python
pop_df .columns

```

```python
pop_df.index
```

```python
#경계정보를 가진 json파일
geo_data=json.load(open('./data/경기도행정구역경계.json',encoding='utf-8'))
geo_data
```

```python

pop_df.columns=pop_df.columns.map(str)
#지도생성
gyeonggi_map =g.Map(location=[37.509671,127.055517],
                   zoom_start=10,
                   tiles='Stamen Terrain')



#구분된 경계를 표시할 떄 사용하는 객체
year='2017'
g.Choropleth(geo_data =geo_data,
            data=pop_df[year],
            columns=[pop_df.index,pop_df[year]],
            key_on='feature.properties.name').add_to(gyeonggi_map)

gyeonggi_map
```
